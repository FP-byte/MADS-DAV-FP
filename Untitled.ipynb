```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Create a synthetic dataset
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)
```

```python
# Perform K-means clustering
kmeans = KMeans(n_clusters=3)
y_kmeans = kmeans.fit_predict(X)
```

```python
# Calculate the Silhouette Score
silhouette_avg = silhouette_score(X, y_kmeans)
print(f'Silhouette Score: {silhouette_avg:.2f}')
```

```python
# Plot the clusters and their centers
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('K-means Clustering with Silhouette Score')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# Create a DataFrame to show documents and their cluster labels
df = pd.DataFrame({'Document': documents, 'Cluster': y_kmeans})

# Group documents by cluster
clustered_docs = df.groupby('Cluster')['Document'].apply(list).reset_index()

# Assign meaningful labels to each cluster
cluster_labels = {}
for idx, row in clustered_docs.iterrows():
    # You can analyze the documents in each cluster to assign a label
    # Here we're using a simple heuristic based on common words
    if idx == 0:
        cluster_labels[idx] = "Cat-related"
    else:
        cluster_labels[idx] = "Dog-related"

# Map the cluster labels back to the DataFrame
df['Cluster Label'] = df['Cluster'].map(cluster_labels)
```

```python
# Display the DataFrame with assigned labels
print(df)

# Plot the t-SNE results and K-means clustering
plt.figure(figsize=(10, 6))
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=df['Cluster'], cmap='viridis', s=100, alpha=0.6)
plt.title('t-SNE Visualization with K-means Clustering')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.colorbar(label='Cluster Label')
plt.show()
```
